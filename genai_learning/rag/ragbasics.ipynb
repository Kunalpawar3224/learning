```python
from langchain_community.document_loaders import PyPDFLoader

path = "budget_speech.pdf"

loader = PyPDFLoader(path)
```

```python
docs = loader.load()
```

```python
docs[0].page_content
```

```python
from langchain_text_splitters import CharacterTextSplitter
```

```python
text_splitter = CharacterTextSplitter(chunk_size = 1000)
```

```python
chunks = text_splitter.split_documents(docs)
```

```python
print(chunks[1].page_content)
```

```python
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
embeddings_model_name = "all-MiniLM-L6-v2"
```

```python
embeddings = HuggingFaceEmbeddings(model_name=embeddings_model_name)
```

```python
vector_db = FAISS.from_documents(chunks, embeddings)
```

```python
result = vector_db.similarity_search("Building Rural Prosperity and Resilience")
```

```python
print(result[0].page_content)
```

```python
# from fastapi import FastAPI
from groq import Groq
# from pydantic import BaseModel
# import uvicorn
# from main import chatbot
from env import api_key, llama_model
from prompt import prompt
print("prompt: ", prompt)
```

```python
def chatbot(message):
    # Retrieve relevant documents
    # results = vector_db.similarity_search(message, k=3)
    # context = "\n".join([doc.page_content for doc in results])

    result = vector_db.similarity_search(message)
    context = []

    for i in result:
        context.append(f"Chunk: {i.page_content}")
    # print("context: ", context)
    client = Groq(api_key=api_key)
    completion = client.chat.completions.create(
        model= llama_model,
        messages=[
        {
            "role": "system",
            "content": prompt
        },
        {
            "role": "user",
            "content": f"Context: {context}\n\nQuestion: {message}"
            # "content": message
        }
        ],
        temperature=0,     # it gives random or creative answers, the more the temp the more your answer is different than static or previous one 
        max_completion_tokens=1024,   # It is for how long your output would be or length of the output   
                                      # Approximately 1 token = 0.75 word  i.e 100 tokens = 75 words
        # top_p=1,
        # stream=False,
        # stop=None
    )
    return completion.choices[0].message.content
```
